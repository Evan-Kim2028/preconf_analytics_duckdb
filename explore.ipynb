```python
import polars as pl
from pipe.data_processing import load_and_join_data, read_single_table

pl.Config.set_fmt_str_lengths(200)
pl.Config.set_fmt_float("full")
```

```python
# Define the DuckDB database filename
db_filename = "mev_commit.duckdb"

# Specify the tables you want to read
tables_to_read = ["encrypted_stores", "commit_stores", "commits_processed"]

# Load and join data
commitments_df = load_and_join_data(db_filename, tables_to_read).with_columns(pl.from_epoch('timestamp', time_unit='ms').alias('datetime'))

# Check if DataFrame is loaded
if not commitments_df.is_empty():
    print(f"Commitments DataFrame Shape: {commitments_df.shape}")
    commitments_df.head()
else:
    print("Failed to load commitments data.")
```

```python
commitments_df.filter(pl.col('datetime') > pl.date(2024,10,15)).group_by('commiter').agg(pl.len().alias('count'))
```
